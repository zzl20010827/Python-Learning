# 二维正态分布可视化  
导入包

```python
import numpy as np
import matplotlib.pyplot as plt
# 3d绘图
from mpl_toolkits.mplot3d import Axes3D
# 导入多维正态分布
from scipy.stats import multivariate_normal
```

正态分布参数设置

```python
# 协方差
cov = np.eye(2)
# 均值
mean = np.zeros(2)
```

画图采样点

```python
# 横纵坐标采样数
length = 100
# 准备画布
X, Y = np.meshgrid(np.linspace(-3, 3, length), np.linspace(-3, 3, length))
z = np.dstack((X, Y))
# 调用multivariate_normal
rv = multivariate_normal(mean=mean, cov=cov)
# 计算采样点处的概率密度
Z = rv.pdf(z)
```

画图

```python
# 图像设置
fig = plt.figure(figsize=(12, 8))
# 新增图像
ax = fig.add_axes(Axes3D(fig))
# 绘制图像并展示
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```
